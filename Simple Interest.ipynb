```python
def Simple_Interest(principal,rate,time):
    Simple_interest = principal * rate / 100 * time
    Total_amount = Simple_interest + principal

    return Simple_interest, Total_amount

def main():
    principal = float(input("Enter the principal: "))
    rate = float(input("Enter the rate(%): "))
    time = float(input("Enter the time(years): "))
    print(Simple_Interest(principal,rate,time))

print(main())
    
    
    
```
