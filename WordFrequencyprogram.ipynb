```python
import string
from collections import Counter

def word_frequency(file_name):
    with open(file_name, 'r') as file:
        text = file.read().lower()

        # Remove punctuation
        for char in string.punctuation:
            text = text.replace(char, '')

        words = text.split()

        # Count word frequency
        word_count = Counter(words)

        # Get top 5 words
        top_five = word_count.most_common(5)

        print("Top 5 most frequent words:")
        for word, count in top_five:
            print(word, ":", count)

# Example usage
word_frequency("sample.txt")
```

```python
def check_even(func):
    def wrapper(num):
        if num % 2 == 0:
            return func(num)
        else:
            print("Number is not even")
    return wrapper

@check_even
def display(num):
    print("Even number:", num)

display(4)
display(3)
```
