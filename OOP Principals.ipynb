```python
# class and object
class Student:
    def __init__(self, name):
        self.name = name

    def display(self):
        print("Name:", self.name)

s1 = Student("Lenny")
s1.display()
```

```python
# inheritance
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

d = Dog()
d.speak()
d.bark()
```

```python
# polymorphsim
class Bird:
    def sound(self):
        print("Bird makes sound")

class Parrot(Bird):
    def sound(self):
        print("Parrot talks")

b = Bird()
p = Parrot()

b.sound()
p.sound()
```
