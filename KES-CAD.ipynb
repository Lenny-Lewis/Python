```python
# Function to convert Kenya Shillings to Canadian Dollars
def convert_kes_to_cad(kes):
    rate = 0.0095  # Example exchange rate
    cad = kes * rate
    return cad


# Main program
print("=== KES to CAD Converter ===")

# Get user input
amount_kes = float(input("Enter amount in Kenya Shillings (KES): "))

# Call the function
amount_cad = convert_kes_to_cad(amount_kes)

# Display result
print(f"\nKES {amount_kes} is equal to CAD {amount_cad:.2f}")
```
