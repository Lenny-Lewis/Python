```python
# Function to calculate average marks
def calculate_average(marks_list):
    # Sum all marks in the list
    total = sum(marks_list)
    
    # Calculate average
    average = total / len(marks_list)
    
    # Return the average value
    return average


# Create an empty list to store marks
marks = []

print("Enter marks for five units:\n")

# Loop to collect 5 marks from the user
for i in range(1, 6):
    mark = float(input(f"Enter marks for unit {i}: "))
    marks.append(mark)  # Add each mark to the list

# Call the function to compute the average
avg_marks = calculate_average(marks)

# Display results
print("\n===== RESULTS =====")
print(f"Marks entered: {marks}")
print(f"Average marks: {avg_marks:.2f}")
```
