```python
# Function to compute sum and average of marks
def compute_sum_average(marks):
    """
    This function takes a list of marks and returns:
    - total sum of marks
    - average marks
    """
    total = sum(marks)                 # Sum all marks
    average = total / len(marks)       # Calculate average
    return total, average

# Main program
print("=== KPSEA Student Marks Calculator ===")

# Empty list to store marks
marks_list = []

# Loop to collect 7 unit marks from the user
for i in range(1, 8):
    mark = float(input(f"Enter marks for unit {i}: "))
    marks_list.append(mark)

# Call the function to compute sum and average
total_marks, average_marks = compute_sum_average(marks_list)

# Display results
print("\n===== RESULTS =====")
print(f"Marks entered: {marks_list}")
print(f"Total Marks: {total_marks}")
print(f"Average Marks: {average_marks:.2f}")
```
