```python
def marks():
   
    Math = float(input("Enter Score for Math: "))
    Eng = float(input("Enter Score for Eng: "))
    Sci = float(input("Enter Score for Sci: "))
    Sst = float(input("Enter Score for Sst: "))
    Cre = float(input("Enter Score for Cre: "))

    
    Total_marks = Math +Eng +Sci +Sst +Cre
    Mean_marks = Total_marks / 5

    print(f"The total mark is: {Total_marks}")
    print(f"The mean mark is: {Mean_marks:.2f}")
    

    if Mean_marks >= 80 and Mean_marks <=100:
        print("You have a mean grade of A")
    elif Mean_marks >=60 and Mean_marks <=79:
        print("You have a mean grade of B")
    elif Mean_marks >=50 and Mean_marks <=59:
        print("You have a mean grade of C")
    elif Mean_marks >=40 and Mean_marks <=49:
        print("You have a mean grade of C")
    else:
        print("You are on your own, Marks Invalid")

    
print(marks())

    
    
    
    
    




    
```
