```python
import math
def CylinderVolume(radius, height):
    volume = math.pi * radius * radius * height
    return volume;

def main():
    radius = float(input("Enter the radius of the cylinder: "))
    height = float(input("Enter the height of the cylinder: "))

    print("\n===== THE VOLUME OF THE CYLINDER ====")
    print(CylinderVolume(radius, height))
print(main())  
    
    
```

```python
# Import math module to use the value of pi
import math

# Prompt user to enter the radius
radius = float(input("Enter the radius of the cylinder: "))

# Prompt user to enter the height
height = float(input("Enter the height of the cylinder: "))

# Calculate the volume using the formula: π * r^2 * h
volume = math.pi * (radius ** 2) * height

# Display the result
print("\n===== CYLINDER VOLUME RESULT =====")
print(f"Radius: {radius}")
print(f"Height: {height}")
print(f"Volume of the cylinder is: {volume:.2f} cubic units")
```
