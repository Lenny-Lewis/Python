```python
def calculator():
    print("Simple Calculator")
    print("Operations: +  -  *  /")

    try:
        num1 = float(input("Enter first number: "))
        operator = input("Enter operator (+, -, *, /): ")
        num2 = float(input("Enter second number: "))

        if operator == "+":
            result = num1 + num2
        elif operator == "-":
            result = num1 - num2
        elif operator == "*":
            result = num1 * num2
        elif operator == "/":
            if num2 == 0:
                print("Error: Division by zero is not allowed.")
                return
            result = num1 / num2
        else:
            print("Error: Invalid operator.")
            return

        print("Result:", result)

    except ValueError:
        print("Error: Please enter valid numeric values.")

# Run the calculator
calculator()
```

```python
# lambda functions
# Normal function
def square(x):
    return x * x

# Lambda equivalent
square_lambda = lambda x: x * x
print(square_lambda(5))
```
