```python
# Import pandas
import pandas as pd

# Create a sample dataset
data = {
    "Name": ["A", "B", "C", "D"],
    "Color": ["Red", "Blue", "Green", "Blue"]
}

df = pd.DataFrame(data)

print("Original Dataset:")
print(df)

# Apply dummification (one-hot encoding)
df_dummies = pd.get_dummies(df, columns=["Color"])

print("\nAfter Dummification:")
print(df_dummies)
```
